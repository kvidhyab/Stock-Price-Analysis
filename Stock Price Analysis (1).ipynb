```python
import pandas as pd

# Initialize an empty DataFrame
all_data = pd.DataFrame()

# Concatenate DataFrames from each file in company_list
for file in company_list:
    current_df = pd.read_csv(file)
    all_data = pd.concat([all_data, current_df], ignore_index=True)
```

```python
app = pd.read_csv(company_list[0])
amzn = pd.read_csv(company_list[1])
google = pd.read_csv(company_list[2])
msft = pd.read_csv(company_list[3])
```

```python
closing_price = pd.DataFrame()
```

```python
closing_price['apple_close'] = app['close']
closing_price['amzn_close'] = amzn['close']
closing_price['google_close'] = google['close']
closing_price['msft_close'] = msft['close']
```

```python
closing_price
```

```python
sns.pairplot(closing_price)
```

```python
#  starighter the line higher the correlation
```

```python
closing_price.corr()
```

```python
sns.heatmap(closing_price.corr(), annot= True)
```

```python
# lighter the colour higher the co-relation 
```

### Pair-Grid

```python
closing_price['apple_close']
```

```python
closing_price['apple_close'].shift(1)
```

```python
(closing_price['apple_close'] - closing_price['apple_close'].shift(1))/closing_price['apple_close'].shift(1)
```

```python
for col in closing_price.columns:
       closing_price[col + '_pct_change'] = (closing_price[col] - closing_price[col].shift(1))/closing_price[col].shift(1)
    
```

```python
closing_price
```

```python
closing_price.columns
```

```python
closing_p =  closing_price[['apple_close_pct_change', 'amzn_close_pct_change',
       'google_close_pct_change', 'msft_close_pct_change']]
```

```python
closing_p
```

```python
g = sns.PairGrid(data = closing_p)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
closing_p.corr()
```

Amazon and microsoft have the highest correlation (0.40), which means if my amazon stock price decreases the probability
that my microsoft price also goes down is 40%.
